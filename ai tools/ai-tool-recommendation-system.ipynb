# Creating AI Recommend System Using Natural Language Processing****

```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity


# Remove the 'Review', 'Tool Link', 'Free/Paid/Other', 'Description', and 'Charges' columns
columns_to_remove = ['Review', 'Tool Link', 'Free/Paid/Other', 'Description', 'Charges', 'Major Category']
df = df.drop(columns=columns_to_remove)

# Define a custom function to clean and concatenate values for 'Useable For' column
def clean_and_concat(values):
    cleaned_values = []
    for value in values:
        # Split by '/', strip, and join with ' and ' if it's not empty
        parts = [part.strip() for part in value.split('/') if part.strip()]
        cleaned_value = ' and '.join(parts)
        cleaned_values.append(cleaned_value)
    return cleaned_values

# Apply the custom function to the 'Useable For' column
df['Useable For'] = clean_and_concat(df['Useable For'])

# Reset the index of the DataFrame
df.reset_index(drop=True, inplace=True)

# You can add your own Use Case here
user_useable_for = "audio editing"

# Use TF-IDF Vectorization for NLP
tfidf_vectorizer = TfidfVectorizer(stop_words='english')
tfidf_matrix = tfidf_vectorizer.fit_transform(df['Useable For'])

# Calculate cosine similarity between user input and tool descriptions
user_input_vector = tfidf_vectorizer.transform([user_useable_for])
cosine_similarities = cosine_similarity(user_input_vector, tfidf_matrix)

# Get the indices of the most similar tools
similar_tool_indices = cosine_similarities.argsort()[0][::-1]

# Define a function to suggest tools based on NLP similarity
def suggest_tools_nlp(df, similar_tool_indices, num_suggestions=5):
    top_tools = df.iloc[similar_tool_indices[:num_suggestions]]
    return top_tools

# Call the function to get NLP-based tool suggestions
nlp_suggested_tools = suggest_tools_nlp(df, similar_tool_indices)

# Display the suggested AI Tool Names and corresponding 'Useable For' values
print(nlp_suggested_tools[['AI Tool Name', 'Useable For']])
```
